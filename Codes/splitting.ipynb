```python
import os
import shutil
import random

# Set your dataset folder
dataset_folder = r"C:\Users\Acer\Downloads\Final\Healthy"
output_folder = dataset_folder  # Keeping output in the same parent folder

# Split ratios
train_ratio = 0.7
val_ratio = 0.15
test_ratio = 0.15

# Get all images
image_extensions = (".jpg", ".png", ".jpeg")
images = [f for f in os.listdir(dataset_folder) if f.endswith(image_extensions)]
random.shuffle(images)

# Split the dataset
total = len(images)
train_split = int(total * train_ratio)
val_split = int(total * (train_ratio + val_ratio))

train_images = images[:train_split]
val_images = images[train_split:val_split]
test_images = images[val_split:]

# Define destination folders
folders = {
    "Train": ("Train/images", "Train/labels"),
    "Val": ("Val/images", "Val/labels"),
    "Test": ("Test/images", "Test/labels"),
}

# Create directories
for img_dir, lbl_dir in folders.values():
    os.makedirs(os.path.join(output_folder, img_dir), exist_ok=True)
    os.makedirs(os.path.join(output_folder, lbl_dir), exist_ok=True)

# Function to move files
def move_files(image_list, img_folder, lbl_folder):
    for img_name in image_list:
        img_path = os.path.join(dataset_folder, img_name)
        lbl_path = os.path.join(dataset_folder, os.path.splitext(img_name)[0] + ".txt")
        
        # Move image
        shutil.move(img_path, os.path.join(output_folder, img_folder, img_name))
        
        # Move label (if exists)
        if os.path.exists(lbl_path):
            shutil.move(lbl_path, os.path.join(output_folder, lbl_folder, os.path.basename(lbl_path)))

# Move files to respective folders
move_files(train_images, *folders["Train"])
move_files(val_images, *folders["Val"])
move_files(test_images, *folders["Test"])

print(f"✅ Dataset split complete!")
print(f"Train: {len(train_images)} | Val: {len(val_images)} | Test: {len(test_images)}")
```
